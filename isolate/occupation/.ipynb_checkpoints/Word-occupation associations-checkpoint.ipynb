# Table of associations

Associations are made based on what we've explicitly coded, along with a comprehensive database of synonyms called `wordnet` included in `nltk`.

This code loops through the occupational codes in alphabetical order.
Within each occupational code it loops through explicitly coded words in alphabetical order.
It puts this word in parentheses, and then displays all synonyms associated with that word.

```python
from IPython.core.display import HTML
from collections import Counter

html = ""

# what OCCs have terms coded to them?
occ_codes = set( [x['code'] for x in occ.codes] )

# sort alphabetically
occ_codes = sorted(occ_codes)

for occ_code in occ_codes:
    html += "<p>"
    html += "<b>%s</b>:   " % occ_code
    
    this_occ_codes = list(filter(lambda x: x['code'] == occ_code, occ.codes))
    
    # group by source [which file did the coding come from], by count
    sources = Counter([x['source'] for x in this_occ_codes])
    
    for source in sources:
        html += "(%s) " % source
        
        this_source_codes = filter(lambda x: x['source'] == source, this_occ_codes)
        for code in this_source_codes:
            html += "%s, " % code['term']
    html += "</p>"

HTML(html)
```
